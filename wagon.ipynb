```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read CSV file
data = pd.read_csv("C:/Users/sabarish jayakumar/OneDrive/Desktop/KAD/WagonWheel/data.csv")

# Step 2: Define boundary coordinates for each region
# You need to define these coordinates based on your image
regions = {
    'Region 1': [(64, 261), (134, 96)],
    'Region 2': [(134, 96), (294, 25)],
    'Region 3': [(294, 25), (467, 95)],
    'Region 4': [(467, 95), (536, 263)],
    'Region 5': [(536, 263), (464, 429)],
    'Region 6': [(464, 429), (300, 495)],
    'Region 7': [(300, 495), (130, 425)],
    'Region 8': [(130, 425), (64, 261)]
}

# Step 3: Map wagon wheel data to each region
region_counts = {region: 0 for region in regions}

for index, row in data.iterrows():
    wagon_wheel_position = row['WagonWheel']
    # Map wagon wheel position to region
    if wagon_wheel_position == 'Point':
        region_counts['Region 1'] += 1
    elif wagon_wheel_position == 'Thirdman':
        region_counts['Region 2'] += 1
    elif wagon_wheel_position == 'Fineleg':
        region_counts['Region 3'] += 1
    elif wagon_wheel_position == 'Squareleg':
        region_counts['Region 4'] += 1
    elif wagon_wheel_position == 'Midwicket':
        region_counts['Region 5'] += 1
    elif wagon_wheel_position == 'Long-on':
        region_counts['Region 6'] += 1
    elif wagon_wheel_position == 'Long-off':
        region_counts['Region 7'] += 1
    elif wagon_wheel_position == 'Covers':
        region_counts['Region 8'] += 1

# Step 4: Calculate percentage of balls in each region
total_balls = len(data)
region_percentages = {region: (count / total_balls) * 100 for region, count in region_counts.items()}

# Step 5: Plot wagon wheel with percentages
img = plt.imread("C:/Users/sabarish jayakumar/OneDrive/Desktop/KAD/WagonWheel/wagon.jpg")
plt.imshow(img)

for region, boundary in regions.items():
    x_center = (boundary[0][0] + boundary[1][0]) / 2
    y_center = (boundary[0][1] + boundary[1][1]) / 2
    plt.text(x_center, y_center, f"{region_percentages[region]:.2f}%", ha='center', va='center', fontsize=12, color='white')

plt.axis('off')
plt.show()
```
